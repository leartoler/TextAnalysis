# Importar

```
content = open("Critica.txt").read()
```

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

# Procesamiento

```
#Lo siguiente es para hacer todas las palabras iguales, y remover puntos y demás
import string
for punc_char in string.punctuation:
  content = content.replace(punc_char,'') #remueve las puntuaciones
content = content.lower()
content = ''.join([char for char in content if char in ' qwertyuiopasdfghjklñzxcvbnm'])
```

```
plt.figure(figsize=(20,9))
wordcloud = WordCloud(max_font_size = 50, max_words=1000,background_color="white").generate(content)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
```

```
#Se cambia el color. En HSL. Para este caso es como un color azuloso
import random
def color_func(**kwargs):
  return "hsl(191, {}%, 25%)".format(random.randint(0,50))
```

```
#Se aplica el color
plt.figure(figsize=(20,9))
wordcloud = WordCloud(max_font_size=50, max_words=1000, background_color="white").generate(content)
default_colors=wordcloud.to_array() #Recolorea
plt.imshow(wordcloud.recolor(color_func=color_func),interpolation="bilinear") #recolorea usando la funcion de arriba
plt.axis("off")
```

# Estilo

A continuación, si bien ya se cuenta con la nube de palabras, lo siguiente es para que se acomoden las palabras dada una imagen. Es importante que la imagen que sirva de máscara tenga sólo dos colores, y con fondo blanco.

```
#Se crea una máscara para poner la nube
import numpy as np
from PIL import Image
```

```
mask = np.array(Image.open("/content/Kant.png"))
plt.figure(figsize=(20,9))
wordcloud = WordCloud(max_font_size=30, max_words=1000, 
                      mask=mask,
                      background_color="pink").generate(content)
default_colors = wordcloud.to_array()
plt.imshow(wordcloud.recolor(color_func=color_func),interpolation="bilinear") #recolorea usando la funcion de arriba
plt.axis("off")



```
